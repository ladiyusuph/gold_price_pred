```python
#Importing the dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics 
```

## Data Collection and Preprocessing

```python
#Reading the csv files
gold_df = pd.read_csv('gld_price_data.csv')
gold_df.head()
```

```python
#Viewing the end of the dataframe
gold_df.tail()
```

```python
#Viewing the no of rows and columns 
gold_df.shape
```

```python
#Checking for missing values in the dataframe
gold_df.isnull().sum()
```

```python
#Additional information about the columns in dataframe
gold_df.info()
```

```python
#Viewing statistical summary of the dataframe
gold_df.describe().transpose()
```

```python
#Viewing the correlation between the different variables in the dataframe

corr = gold_df.corr()
plt.figure(figsize = (7,6))
sns.heatmap(corr, annot = True, annot_kws = {'size':8}, cbar = True, square = True, fmt = '.1f', cmap = 'Blues')
```

#### Gold is Positively correlated with Silver, therefore the more Gold prices increase the more Silver will increase

```python
#viewing the corr values of GLD
print(corr['GLD'])
```

```python
#Check the distribution of the Gold Column

sns.distplot(gold_df['GLD'], color = 'red')
```

```python
#Splitting the data into features and label
X = gold_df.drop(columns = ['Date', 'GLD'], axis = 1)
y = gold_df.GLD
```

```python
#Viewing the shape of the separation
print(X.shape, y.shape)
```

```python
#Split the data into training and test data
SEED = 345
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = SEED)
```

```python
#Validataing the split
print(X.shape, X_train.shape, X_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

## Model Building

```python
#Initiating a Random Forest Model

rf = RandomForestRegressor(n_estimators = 100)
```

```python
#Training the model 
rf.fit(X_train, y_train)
```

## Model Evaluation

```python
#Prediction on test data
y_test_pred = rf.predict(X_test)

#RSquared error
err = metrics.r2_score(y_test, y_test_pred)

print(f"The RSquared error is: {err}")
```

```python
y_test = list(y_test)
#Visualizing the predicted and actual values

plt.plot(y_test, color = 'blue', label = 'Actual Values')
plt.plot(y_test_pred, color = 'green', label = 'Predicted Values')
plt.title('Actual VS Predicted Price')
plt.xlabel('Number of Values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
